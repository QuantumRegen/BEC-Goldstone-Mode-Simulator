```python
import numpy as np
from scipy.fftpack import fft, ifft
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ────────────────────────────────────────────────
# Global / default parameters & fallback grid
# ────────────────────────────────────────────────
dt = 0.01
g = 100.0
mu = 10.0

barrier_height = 20.0
barrier_width = 1.0
well_sep = 5.0

# Default (normal) resolution grid & potential (fallback)
N = 1024
dx = 0.1
x = dx * (np.arange(N) - N//2)
k = 2 * np.pi * np.fft.fftfreq(N, d=dx)
k2 = k**2 / 2

V = barrier_height * np.exp(-x**2 / (2 * barrier_width**2))
V -= 5.0 * np.exp(-(x - well_sep/2)**2 / 2)          # left
V -= 4.8 * np.exp(-(x + well_sep/2)**2 / 2)          # right (bias)

# ────────────────────────────────────────────────
# Fit functions
# ────────────────────────────────────────────────
def double_gauss(x, A1, mu1, sigma1, A2, mu2, sigma2):
    return A1 * np.exp(-(x-mu1)**2 / (2*sigma1**2)) + A2 * np.exp(-(x-mu2)**2 / (2*sigma2**2))

def single_gauss(x, A, mu, sigma):
    return A * np.exp(-(x-mu)**2 / (2*sigma**2))

p0_double = [0.1, -2.5, 1.0, 0.1, 2.5, 1.0]
p0_single = [0.2, 0.0, 2.0]

# ────────────────────────────────────────────────
# GPE imaginary-time step
# ────────────────────────────────────────────────
def gpe_imag_time_step(psi, V, g, dt, k2, mu_I=0.0):
    V_eff = V - (mu_I**2 * 20.0) * np.abs(psi)**2
    psi = np.exp(-dt / 2 * (V_eff + g * np.abs(psi)**2 - mu)) * psi
    psi_k = fft(psi)
    psi_k = np.exp(-dt * k2) * psi_k
    psi = ifft(psi_k)
    psi = np.exp(-dt / 2 * (V_eff + g * np.abs(psi)**2 - mu)) * psi
    norm = np.sqrt(np.sum(np.abs(psi)**2) * dx)  # note: dx is global fallback
    psi /= norm
    return psi

# ────────────────────────────────────────────────
# Goldstone noise test (add small phase noise → re-evolve → fidelity)
# ────────────────────────────────────────────────
def goldstone_noise_test(psi, V, g, dt, k2, mu_I, noise_amp=0.1, steps=50):
    phase_noise = noise_amp * np.random.randn(len(psi))
    psi_noisy = psi * np.exp(1j * phase_noise)
    
    for _ in range(steps):
        psi_noisy = gpe_imag_time_step(psi_noisy, V, g, dt, k2, mu_I)
    
    fidelity = np.abs(np.sum(np.conj(psi) * psi_noisy) * dx)**2
    return psi_noisy, fidelity

# ────────────────────────────────────────────────
# Main sweep
# ────────────────────────────────────────────────
mu_I_values = [0.0, 1.0, 1.4, 1.6, 1.8, 2.0, 2.2, 3.0]
print("\n=== Full Sweep: Asym wells, Phase Zoom, Variance, Hi-Res, Goldstone Test ===")

for mu_I_val in mu_I_values:
    print(f"\nμ_I = {mu_I_val:.1f}")
    
    # Higher res for high μ_I (to resolve sharp spikes)
    if mu_I_val >= 2.0:
        N_local = 2048
        dx_local = 0.05
        x_local = dx_local * (np.arange(N_local) - N_local//2)
        k_local = 2 * np.pi * np.fft.fftfreq(N_local, d=dx_local)
        k2_local = k_local**2 / 2
        
        V_local = barrier_height * np.exp(-x_local**2 / (2 * barrier_width**2))
        V_local -= 5.0 * np.exp(-(x_local - well_sep/2)**2 / 2)
        V_local -= 4.8 * np.exp(-(x_local + well_sep/2)**2 / 2)
        
        print(f"(Using higher res: N={N_local}, dx={dx_local} pm)")
    else:
        N_local = N
        dx_local = dx
        x_local = x
        V_local = V
        k2_local = k2
    
    # Initial condition
    psi = np.exp(-x_local**2 / 20) / np.pi**0.25
    psi /= np.sqrt(np.sum(np.abs(psi)**2) * dx_local)
    
    # More steps for high μ_I
    steps = 200 if mu_I_val >= 2.0 else 120
    for _ in range(steps):
        psi = gpe_imag_time_step(psi, V_local, g, dt, k2_local, mu_I=mu_I_val)
    
    density = np.abs(psi)**2
    total_n = np.trapezoid(density, x_local)
    max_d = np.max(density)
    print(f"Total density: {total_n:.6f}   Max density: {max_d:.6f}")
    
    # Phase & variance in high-density region (>10% max)
    phase = np.unwrap(np.angle(psi))
    high_dens_mask = density > 0.1 * max_d
    phase_var = np.var(phase[high_dens_mask]) if np.any(high_dens_mask) else np.nan
    print(f"Phase variance (high-density region): {phase_var:.4f} rad²")
    
    # Full phase plot
    plt.figure(figsize=(8,4))
    plt.plot(x_local, phase, 'g-')
    plt.xlabel('x (pm)'); plt.ylabel('Phase (rad)')
    plt.title(f'Phase at μ_I = {mu_I_val:.1f}')
    plt.grid(True, alpha=0.3)
    plt.savefig(f'phase_muI_{mu_I_val:.1f}.png', dpi=150, bbox_inches='tight')
    plt.close()
    
    # Zoom phase (|x| < 10 pm)
    zoom_mask = np.abs(x_local) < 10
    if np.any(zoom_mask):
        plt.figure(figsize=(6,4))
        plt.plot(x_local[zoom_mask], phase[zoom_mask], 'g-')
        plt.xlabel('x (pm)'); plt.ylabel('Phase (rad)')
        plt.title(f'Phase zoom (|x|<10 pm) at μ_I = {mu_I_val:.1f}')
        plt.grid(True, alpha=0.3)
        plt.savefig(f'phase_zoom_muI_{mu_I_val:.1f}.png', dpi=200, bbox_inches='tight')
        plt.close()
    
    # Density fits
    fit_success = False
    try:
        popt_d, _ = curve_fit(double_gauss, x_local, density, p0=p0_double, maxfev=8000)
        sep = abs(popt_d[1] - popt_d[4])
        print(f"Double fit: A1/A2 = {popt_d[0]:.4f}/{popt_d[3]:.4f}   sep = {sep:.4f} pm")
        fit_success = True
    except:
        print("Double Gaussian fit failed")
    
    if not fit_success or sep < 7.0:
        try:
            popt_s, _ = curve_fit(single_gauss, x_local, density, p0=p0_single, maxfev=5000)
            print(f"Single Gaussian fit: A = {popt_s[0]:.4f}, center = {popt_s[1]:.4f}, σ = {popt_s[2]:.4f} pm")
        except:
            print("Single Gaussian fit failed")
    
    # Density plot
    plt.figure(figsize=(8,5))
    plt.plot(x_local, density, 'b-', label='Density')
    try:
        plt.plot(x_local, double_gauss(x_local, *popt_d), 'r--', label='Double fit')
    except:
        pass
    try:
        plt.plot(x_local, single_gauss(x_local, *popt_s), 'm--', label='Single fit')
    except:
        pass
    plt.xlabel('x (pm)'); plt.ylabel('|ψ|²')
    plt.title(f'Density at μ_I = {mu_I_val:.1f} (asym wells)')
    plt.legend(); plt.grid(True, alpha=0.3)
    plt.savefig(f'density_muI_{mu_I_val:.1f}.png', dpi=150, bbox_inches='tight')
    plt.close()
    
    # Goldstone noise test at transition points
    if mu_I_val in [1.8, 2.0]:
        psi_noisy, fidelity = goldstone_noise_test(psi, V_local, g, dt, k2_local, mu_I_val, noise_amp=0.1, steps=50)
        print(f"Goldstone noise test: Fidelity after noise + 50 steps re-evolve: {fidelity:.4f}")
        # Save noisy relaxed density
        plt.figure(figsize=(8,5))
        plt.plot(x_local, np.abs(psi_noisy)**2, 'c-', label='After noise + relaxation')
        plt.xlabel('x (pm)'); plt.ylabel('|ψ|²')
        plt.title(f'Noisy density after relaxation at μ_I = {mu_I_val:.1f}')
        plt.legend(); plt.grid(True, alpha=0.3)
        plt.savefig(f'noise_relax_muI_{mu_I_val:.1f}.png', dpi=150, bbox_inches='tight')
        plt.close()

print("\nSweep complete. Check PNGs (phase_zoom_*, density_*, noise_relax_*) and phase variance values.")
```
